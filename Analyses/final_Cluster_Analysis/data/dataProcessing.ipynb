```python
# Libraries used
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings # ignore warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("uncleanedData/plus2.csv", index_col='Fact')
data2=data.iloc[0:66, :]
data3=data2.T
```

```python
data3['Banned or not'] = [0,0]
data3
```

```python
# Loading in Census Data
df = pd.read_csv("uncleanedData/CensusData.csv")
df.rename(columns={"Fact": "County Name"}, inplace=True)
df
```

```python
dff = df.set_index('County Name')
dff
```

```python
concatDf = pd.concat([dff, data3])
concatDf
```

```python
concatDf2=concatDf.loc[:, :'Banned or not']
concatDf2
```

```python
leftDf = concatDf.loc[:, 'Banned or not':]
leftDf
```

```python
leftDf.columns
```

```python
concatDf2['Population Estimates, July 1 2021, (V2021)']['Summit County, Ohio']= leftDf.loc['Summit County, Ohio','Population Estimates, July 1, 2021, (V2021)']
concatDf2['Population Estimates, July 1 2021, (V2021)']['Van Wert County, Ohio']= leftDf.loc['Van Wert County, Ohio','Population Estimates, July 1, 2021, (V2021)']
concatDf2['Persons without health insurance, under age 65 years, percent']['Summit County, Ohio']= leftDf.loc['Summit County, Ohio','Persons  without health insurance, under age 65 years, percent']
concatDf2['Persons without health insurance, under age 65 years, percent']['Van Wert County, Ohio']= leftDf.loc['Van Wert County, Ohio','Persons  without health insurance, under age 65 years, percent']
```

```python
concatDf2
```

```python
# Dropped all NA values
df2=concatDf2.dropna()
df2
```

```python
# Removed columns that have records not from 2021
df3 = df2.drop(['Population estimates base, April 1, 2020, (V2021)', 
          'Population, percent change - April 1, 2020 (estimates base) to July 1, 2021, (V2021)', 
          'Population, Census, April 1, 2020', 'Population, Census, April 1, 2010'], axis=1)

df3.head()
```

```python
# remove all commas
# remove all '%'
# remove all '$'
df4 = df3.replace([',', '%', '$'],'', regex=True)
df4.head()
```

```python
# Remove all data that do not anything from 2020/2021/2022
dropyearList = []
for col in df4.columns:
    if '2010' in col:
        dropyearList.append(col)
    elif 'year 2017' in col:
        dropyearList.append(col)
    elif ', 2017 ' in col:
        dropyearList.append(col)
    elif 's, 2019' in col:
        dropyearList.append(col)

df5 = df4.drop(dropyearList, axis=1)
df6 = df5.drop('Total employment, percent change, 2019-2020', axis=1)
df6.head()
```

```python
# replace 'S' and 'Z' with NA values
for col in df6.columns:
    if 'Z' or 'S' in df6[col]:
        df6[col] = df6[col].replace(['Z', 'S'], 0) 
df7 = df6.replace({'\$': ''}, regex=True)

df7.head()
```

```python
# Changes data type object to int
df7.iloc[:, :] = df7.iloc[:, :].apply(pd.to_numeric, errors='coerce')
```

```python
# Converts percentages into Percent Portortion 
percent_cols = df7.filter(like='percent').columns

df7[percent_cols] = df7[percent_cols].apply(lambda x: round(x * df7['Population Estimates, July 1 2021, (V2021)']/100,2))
```

```python
df7.columns= df7.columns.str.replace(", percent", "") # We removed the word percent 
df7
```

```python
# Loads newly cleaned dataset into a new .csv file.
df8 = df7.reset_index()
df9 = df8.rename(columns={"index": "County Name"})
df9.fillna(0, inplace=True)
```

```python
df9.head()
```

```python
df9.to_csv('cleanData/fullData.csv')
```

## Splitting into subsets 
Because of how large this dataset is, we decieded to create subsets of the data that each will include information about Population, housing, and Education/work

### DF subset 1: Population 

```python
dfPop = df9.iloc[:,0:16]
dfPop['Banned or not'] = df9['Banned or not']
dfPop.columns= dfPop.columns.str.replace(", percent", "") # We removed the word percent 
dfPop.columns= dfPop.columns.str.replace("(V2021)", "", regex=True) # ne
dfPop.columns= dfPop.columns.str.replace(', ()', "", regex=True)
```

```python
dfPop.head()
```

```python
dfPop.to_csv('cleanData/popData.csv')
```

### DF subset 2: Housing 

```python
dfHousing = df9.iloc[:, 16:29]
dfHousing['Banned or not'] = df9['Banned or not']
dfHousing.columns= dfHousing.columns.str.replace(", percent", "")
dfHousing.insert(0, 'County Name', df9['County Name'])
```

```python
dfHousing.head()
```

```python
dfHousing.to_csv('cleanData/housingData.csv')
```

### DF subset 3: Edu/Work 

```python
dfEduWork = df9.iloc[:, 30:]
# dfEduWork.columns = dfEduWork.columns.str.replace(", 2017-2021", "")
dfEduWork.columns= dfEduWork.columns.str.replace(", percent", "")
dfEduWork.drop(columns='Total retail sales per capita, 2017')
dfEduWork.insert(0, 'County Name', df9['County Name'])
```
