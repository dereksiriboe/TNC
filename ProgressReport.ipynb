```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.DataFrame({
      'Task': ['Data collection', 'Demographics in general','Renewable energy impact on the economy', 'Progress Report I', 'Methodology selection for data processing analyzing','Reason for full or partial ban', 'Establishment of \'renewable energy exclusion zones\'','Progess Report II','Conclusion','Final Executive Summary/ Final Technical Report'],
      'Start': ['2023-01-16','2023-01-16', '2023-02-23', '2023-03-03','2023-03-06','2023-03-06', '2023-03-27', '2023-04-04', '2023-04-10','2023-04-24'],
      'Finish': ['2023-03-05', '2023-02-23','2023-03-19', '2023-03-10', '2023-03-12','2023-03-26', '2023-04-09','2023-04-11','2023-04-30','2023-05-11'],
      'Category':['Finished', 'Finished','Still working', 'Finished','Still working','Undone','Undone','Undone','Undone','Undone']
})
```

```python
df['Start'] = pd.to_datetime(df['Start'])
df['Finish'] = pd.to_datetime(df['Finish'])
```

```python
colors = {'Finished': 'green', 'Still working': 'yellow', 'Undone': 'red'}
fig = px.timeline(df, x_start="Start", x_end="Finish", y="Task", color="Category", color_discrete_map=colors)
fig.show()
```

```python
fig.update_layout(
    title="Progress Chart",
    width=1250, # set the width to 800 pixels
    height=500 # set the height to 600 pixels
)
```
